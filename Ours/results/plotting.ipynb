```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0.8298, 0.8316, 0.8349, 0.8381, 0.8405, 0.8364])
plt.plot(x, linestyle='--', marker='o', color='b')
plt.ylim(0.8, 0.85) 
xt = ['64', '128', '256', '512', '1024', '2048']
plt.xticks(range(len(xt)), xt, size='large')
plt.xlabel('Embedding dimension')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Embedding dimension')
[plt.axhline(y=i, linestyle='--', c = "k") for i in np.linspace(0.80, 0.85, 6)]
```
